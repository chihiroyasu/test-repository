```python
import random
import time

print("フラッシュ暗算ゲーム!!")
print("※数字は半角で打ってください")
nani = input("足し算か掛け算かやりたい方を打ってください")
nani = str(nani)

if nani == "掛け算":
    
    
    nando = input("難しい・普通・簡単の中から難易度を一つ打ってください")
    nando = str(nando)
    
    if nando == "難しい":

        data=[]
        for i in range(5):
            atai=random.randint(1,20)
            data.append(atai)
            print(atai)
            time.sleep(1)
        kotae = data[0]*data[1]*data[2]*data[3]*data[4]
        answere = input("答えは？")
        answere = int(answere)
        
        if kotae == answere:
            print("おめでとう！正解です！！")
        else:
            print("残念…。不正解(´∵｀)")
            kotae = str(kotae)
            print("答えは" + kotae + "でした")
    
    if nando == "普通":

        data=[]
        for i in range(5):
            atai=random.randint(1,10)
            data.append(atai)
            print(atai)
            time.sleep(1)
        kotae = data[0]*data[1]*data[2]*data[3]*data[4]
        answere = input("答えは？")
        answere = int(answere)
        if kotae == answere:
            print("おめでとう！正解です！！")
        else:
            print("残念…。不正解(´∵｀)")
            kotae = str(kotae)
            print("答えは" + kotae + "でした")
            
    if nando == "簡単":

        data=[]
        for i in range(5):
            atai=random.randint(1,10)
            data.append(atai)
            print(atai)
            time.sleep(2)
        kotae = data[0]*data[1]*data[2]*data[3]*data[4]
        answere = input("答えは？")
        answere = int(answere)
        if kotae == answere:
            print("おめでとう！正解です！！")
        else:
            print("残念…。不正解(´∵｀)")
            kotae = str(kotae)
            print("答えは" + kotae + "でした")

if nani =="足し算":
    
    nando = input("難しい・普通・簡単の中から一つ難易度を打ってください")
    nando = str(nando)

    if nando == "難しい":

        data=[]
        for i in range(5):
            atai=random.randint(1,100)
            data.append(atai)
            print(atai)
            time.sleep(0.5)
        kotae = data[0]+data[1]+data[2]+data[3]+data[4]
        answere = input("答えは？")
        answere = int(answere)
        if kotae == answere:
            print("おめでとう！正解です！！")
        else:
            print("残念…。不正解(´∵｀)")
            kotae = str(kotae)
            print("答えは" + kotae + "でした")

    if nando == "普通":

        data=[]
        for i in range(5):
            atai=random.randint(1,100)
            data.append(atai)
            print(atai)
            time.sleep(1)
        kotae = data[0]+data[1]+data[2]+data[3]+data[4]
        answere = input("答えは？")
        answere = int(answere)
        if kotae == answere:
            print("おめでとう！正解です！！")
        else:
            print("残念…。不正解(´∵｀)")
            kotae = str(kotae)
            print("答えは" + kotae + "でした")
            
    if nando == "簡単":

        data=[]
        for i in range(5):
            atai=random.randint(1,100)
            data.append(atai)
            print(atai)
            time.sleep(3)
        kotae = data[0]+data[1]+data[2]+data[3]+data[4]
        answere = input("答えは？")
        answere = int(answere)
        if kotae == answere:
            print("おめでとう！正解です！！")
        else:
            print("残念…。不正解(´∵｀)")
            kotae = str(kotae)
            print("答えは" + kotae + "でした")
```
